# Venn
This script is for drawing venn diagram.

```python
#!pip install matplotlib-venn

# import library
import matplotlib.pyplot as plt
from matplotlib_venn import venn2,venn2_circles

import pandas as pd
```

```python
data = {
    "animal":["Mouse","Goat","Chicken","Canine", "GuineaPig"],
    "Depleted": [93,51,58,35,28],
    "UnDepleted": [16,10,28,10,8],
    "overlap":[182,74,131,91,88]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
```

```python
plt.rcParams.update({'font.size': 18})
plt.rcParams["font.family"] = "arial"
# loop through the rows using iterrows()
for index, row in df.iterrows():
    depleted = int(row['Depleted'])
    undepleted = int(row['UnDepleted'])
    overlap = int(row['overlap'])
    total = depleted+undepleted+overlap
    depleted_all =  depleted+overlap
    undepleted_all = undepleted+overlap
    depleted_percentage = depleted/depleted_all
    undepleted_percentage = undepleted/undepleted_all
    venn = venn2(subsets = (row['UnDepleted'], row['Depleted'], row['overlap']), 
            set_labels = ('Undepleted ('+str(undepleted_all)+')', 'Kit-Depleted ('+str(depleted_all)+')'),
            set_colors=("#4174ff","#fff603"), #blue, yellow
            subset_label_formatter=lambda x: f"{(x/total):1.1%}",
            alpha = 0.4
                )
    plt.title(row['animal'])
    
    for text in venn.set_labels:
       text.set_fontsize(18)

    for text in venn.subset_labels:
        text.set_fontsize(14)
    
    # add outline
    venn2_circles(subsets=(row['UnDepleted'], row['Depleted'], row['overlap']), linewidth=0.1) 
    plt.savefig(row['animal']+'.svg', format = 'svg', dpi=300)
    plt.show()
```
